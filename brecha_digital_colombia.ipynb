### 1. ¿Qué porcentaje de viviendas en Colombia tienen acceso a Internet?


```python
# Dejamos solo las columnas 'DIRECTORIO', 'P1_DEPARTAMENTO', 'SECUENCIA_ENCUESTA' y 'SECUENCIA_P' en df_v1
df_v1 = df_v1[['DIRECTORIO', 'P1_DEPARTAMENTO', 'CLASE']]
df_v1.head()
```

```python
#unimos la tabla de "Datos vivienda" con la tabla de "Codigos departamentos"
df_v1 = df_v1.merge(
    df_codigos[['Código Departamento', 'Nombre Departamento']].drop_duplicates(),
    how="left",
    left_on="P1_DEPARTAMENTO",
    right_on="Código Departamento"
)

#  Renombramos la columna resultante
df_v1 = df_v1.rename(columns={"Nombre Departamento": "departamento"})

# Mostrar primeras filas para verificar
display(df_v1.head())

# Verificamos si algunas filas quedaron sin asignar departamento
print("Cantidad de filas sin departamento asignado:")
print(df_v1['departamento'].isna().sum())
```

```python
# Mostrar valores únicos de la columna 'departamento'
print(df_v1['departamento'].unique())
```

```python
#empezamos a explorar la tabla H3 y traemos la columna P1075 que corresponde a la pregunta de la encuenta "Tiene internet?" Respuestas si(1), No(2)

#Revisamos las columas de la tabla "Condiciones de vida del hogar"
print(df_h3.columns)

# Seleccionar solo columnas necesarias
df_h3_internet = df_h3[['DIRECTORIO', 'P1075']].copy()

# Si hay duplicados de DIRECTORIO, tomar el primero
df_h3_internet = df_h3_internet.drop_duplicates(subset=['DIRECTORIO'])

# Hacer merge con df_v1
df_v1 = df_v1.merge(
    df_h3_internet,
    how="left",
    on="DIRECTORIO"
)

# Reemplazamos valores numéricos por "Sí" y "No"
df_v1['internet'] = df_v1['P1075'].map({
    1: "Sí",
    2: "No"
})


# Mostrar primeras filas para verificar
display(df_v1.head())
```

```python
#Procesamiento para graficar porcentaje de viviendas conectados a Internet

#Contar hogares por estado de conexión a internet
conteo_internet = df_v1['internet'].value_counts().reset_index()
conteo_internet.columns = ['Internet', 'Cantidad']

# Calcular el porcentaje
total_hogares = conteo_internet['Cantidad'].sum()
conteo_internet['Porcentaje'] = (conteo_internet['Cantidad'] / total_hogares * 100).round(2)

# Estilo
sns.set_theme(style="whitegrid")

# Gráfico de barras por porcentaje
plt.figure(figsize=(6, 4))
ax = sns.barplot(
    data=conteo_internet,
    x='Internet',
    y='Porcentaje',
    hue='Internet',
    dodge=False,
    palette='viridis'
)

# Etiquetas con porcentaje encima de cada barra
for i, row in conteo_internet.iterrows():
    ax.text(i, row['Porcentaje'] + 1, f"{row['Porcentaje']}%", ha='center', fontsize=12)

# Título y ejes
plt.title("Porcentaje de viviendas con conexión a Internet", fontsize=14)
plt.xlabel("¿El hogar tiene conexión a Internet?")
plt.ylabel("Porcentaje de hogares (%)")
plt.tight_layout()
plt.show()
```

### 2. ¿Cuáles son los departamentos con menor y mayor acceso a Internet?

```python
#Procesamiento para graficar acceso a Internet de viviendas por departamento

# Contar Viviendas por departamento y estado de internet
conteo_dep_internet = df_v1.groupby(['departamento', 'internet']).size().reset_index(name='Cantidad')

# Calcular total por departamento
totales_dep = conteo_dep_internet.groupby('departamento')['Cantidad'].transform('sum')

# Calcular porcentaje
conteo_dep_internet['Porcentaje'] = (conteo_dep_internet['Cantidad'] / totales_dep * 100).round(2)

# Ordenar departamentos alfabéticamente
conteo_dep_internet = conteo_dep_internet.sort_values('departamento')

# Estilo
sns.set_theme(style="whitegrid")

# Gráfico de barras agrupadas por porcentaje
plt.figure(figsize=(18, 8))
ax = sns.barplot(
    data=conteo_dep_internet,
    x='departamento',
    y='Porcentaje',
    hue='internet',
    palette="Set2"
)

# Mostrar porcentaje encima de cada barra
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='edge', fontsize=9, padding=2)

# 8. Ajustes estéticos
plt.xticks(rotation=60, ha='right', fontsize=10)
plt.title("Porcentaje de hogares con acceso a Internet por departamento", fontsize=16, weight='bold')
plt.xlabel("Departamento", fontsize=12)
plt.ylabel("Porcentaje de hogares (%)", fontsize=12)
plt.legend(title="¿Tiene Internet?", fontsize=10, title_fontsize=12)
plt.tight_layout()
plt.show()
```

### 2.1. Sacando el Top 5 departamentos con mayor y menos acceso a Internet

```python
# Contar hogares por departamento y estado de internet
conteo = df_v1.groupby(['departamento', 'internet']).size().reset_index(name='Cantidad')

# Pivotear la tabla para tener columnas separadas
pivot = conteo.pivot(index='departamento', columns='internet', values='Cantidad').fillna(0)

# Renombrar columnas si es necesario
pivot.columns.name = None
pivot = pivot.rename(columns={"Sí": "Con_Internet", "No": "Sin_Internet"})

# Calcular total y porcentaje
pivot["Total_Hogares"] = pivot["Con_Internet"] + pivot["Sin_Internet"]
pivot["Porcentaje_Con_Internet"] = (pivot["Con_Internet"] / pivot["Total_Hogares"]) * 100

# Ordenar por porcentaje de acceso
pivot_sorted = pivot.sort_values("Porcentaje_Con_Internet", ascending=False)

# Top 5 con mayor acceso
top_5_mayor = pivot_sorted.head(5)

# Top 5 con menor acceso
top_5_menor = pivot_sorted.tail(5)

# Mostrar resultados
print("TOP 5 DEPARTAMENTOS CON MAYOR ACCESO A INTERNET:")
print(top_5_mayor[["Porcentaje_Con_Internet"]])

print("\nTOP 5 DEPARTAMENTOS CON MENOR ACCESO A INTERNET:")
print(top_5_menor[["Porcentaje_Con_Internet"]])
```

```python
top_5_mayor = top_5_mayor.copy()
top_5_menor = top_5_menor.copy()

# Concatenar top 5 mayor y menor acceso, etiquetando el grupo
top_5_mayor["Grupo"] = "Mayor acceso"
top_5_menor["Grupo"] = "Menor acceso"
df_ranking = pd.concat([top_5_mayor, top_5_menor])

# Ordenar por porcentaje
df_ranking = df_ranking.sort_values("Porcentaje_Con_Internet", ascending=False)

# Configurar estilo
sns.set_theme(style="whitegrid")

# Crear figura
plt.figure(figsize=(10,6))
ax = sns.barplot(
    data=df_ranking,
    y=df_ranking.index,
    x="Porcentaje_Con_Internet",
    hue="Grupo",
    palette="Set2"
)

# Añadir etiquetas de porcentaje
for p in ax.patches:
    width = p.get_width()
    ax.text(
        width + 1,
        p.get_y() + p.get_height() / 2,
        f"{width:.1f}%",
        va="center",
        fontsize=10
    )

# Títulos y etiquetas
plt.title("TOP 5 Departamentos con mayor y menor acceso a Internet", fontsize=16, weight="bold")
plt.xlabel("Porcentaje de hogares con Internet")
plt.ylabel("Departamento")
plt.legend(title="Grupo")

# Ajustar diseño
plt.tight_layout()
plt.show()
```

### 3. ¿Cuál es la cantidad de viviendas que pagan por el servicio de internet fijo, cuántos por internet móvil?

```python
# Usaremos las columnas P1075S1 QUE HACE REFERENCIA A " Internet FIJO"  1 PARA SI Y 2 PARA NO   y la P1075S2 INTERNET MOVIL  1 PARA SI Y 2 PARA NO 

# Seleccionar las columnas necesarias de df_h3
df_internet_detalle = df_h3[['DIRECTORIO', 'P1075S1', 'P1075S2']].copy()

# Eliminar duplicados por si hay varios registros por DIRECTORIO
df_internet_detalle = df_internet_detalle.drop_duplicates(subset=['DIRECTORIO'])

# Unir al df maestro
df_v1 = df_v1.merge(
    df_internet_detalle,
    how='left',
    on='DIRECTORIO'
)

# Mapear los valores 1/2 a texto para mayor claridad
df_v1['internet_fijo'] = df_v1['P1075S1'].map({1: 'Sí', 2: 'No'})
df_v1['internet_movil'] = df_v1['P1075S2'].map({1: 'Sí', 2: 'No'})

# Eliminar las columnas originales si ya no las quieres
df_v1 = df_v1.drop(columns=['P1075S1', 'P1075S2'])

# Mostrar las primeras filas para verificar
display(df_v1.head())
```

```python
# Crear variables booleanas
df_v1['Tiene_Fijo'] = df_v1['internet_fijo'] == 'Sí'
df_v1['Tiene_Movil'] = df_v1['internet_movil'] == 'Sí'

# Cantidad de viviendas con internet fijo
cantidad_fijo = df_v1['Tiene_Fijo'].sum()

# Cantidad de viviendas con internet móvil
cantidad_movil = df_v1['Tiene_Movil'].sum()

# Cantidad de viviendas que tienen ambos
cantidad_ambos = df_v1[(df_v1['Tiene_Fijo']) & (df_v1['Tiene_Movil'])].shape[0]

# Mostrar resultados
print(f"Cantidad de viviendas con Internet Fijo: {cantidad_fijo:,}")
print(f"Cantidad de viviendas con Internet Móvil: {cantidad_movil:,}")
print(f"Cantidad de viviendas con ambos: {cantidad_ambos:,}")
```

```python
datos_resumen = pd.DataFrame({
    "Tipo de conexión": ["Internet Fijo", "Internet Móvil", "Ambos"],
    "Cantidad": [cantidad_fijo, cantidad_movil, cantidad_ambos]
})

# Calcular total de viviendas (al menos con fijo o móvil o ambos)
total_viviendas = datos_resumen["Cantidad"].max() 

total_viviendas = df_v1.shape[0]

# Calcular porcentaje
datos_resumen["Porcentaje"] = (datos_resumen["Cantidad"] / total_viviendas) * 100

# Configurar estilo
sns.set(style="whitegrid")

# Crear gráfico de barras
plt.figure(figsize=(8,5))
ax = sns.barplot(
    data=datos_resumen,
    x="Tipo de conexión",
    y="Porcentaje",
    palette="pastel"
)

# Añadir etiquetas encima de las barras
for p in ax.patches:
    height = p.get_height()
    ax.text(
        p.get_x() + p.get_width()/2,
        height + 1,
        f"{height:.1f}%",
        ha="center",
        fontsize=12
    )

# Título y etiquetas
plt.title("Proporción de viviendas según tipo de conexión a Internet", fontsize=15, weight="bold")
plt.xlabel("")
plt.ylabel("Porcentaje de viviendas")

# Ajustar y mostrar
plt.tight_layout()
plt.show()
```

```python
# Filtrar solo personas que respondieron la razón por la que no usan Internet
df_no_usa_internet = df_p3[df_p3["Razon_no_usa_internet"].notna()].copy()

# Diccionario de códigos de respuesta
razones_dict = {
    1: "Muy costoso",
    2: "No lo necesita",
    3: "No sabe usarlo",
    4: "No hay cobertura en la zona",
    5: "No le permiten usarlo",
    6: "Seguridad o privacidad",
    7: "Otra razón"
}

# Convertir a entero y mapear
df_no_usa_internet["Razon_no_usa_internet"] = df_no_usa_internet["Razon_no_usa_internet"].astype(int).map(razones_dict)

# Calcular frecuencias y porcentajes
conteo = df_no_usa_internet["Razon_no_usa_internet"].value_counts()
porcentaje = (conteo / conteo.sum() * 100).round(2)

# Crear tabla
resumen_razones_final = pd.DataFrame({
    "Razón": conteo.index,
    "Cantidad de personas": conteo.values,
    "Porcentaje (%)": porcentaje.values
})

# Mostrar tabla
print(resumen_razones_final)
```

```python
#Grafico ¿Cuál es la razón principal por la que las personas no usan Internet?
resumen_razones_final = resumen_razones_final.sort_values("Porcentaje (%)", ascending=False)

# Estilo con fondo blanco y colores variados
sns.set(style="whitegrid")

# Crear gráfico con diferentes colores (por ejemplo Set2)
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x="Porcentaje (%)",
    y="Razón",
    data=resumen_razones_final,
    palette="Paired",
    orient="h"
)

# Etiquetas al final de cada barra
for index, value in enumerate(resumen_razones_final["Porcentaje (%)"]):
    plt.text(value + 0.5, index, f"{value}%", va='center')

# Títulos
plt.title("Razones principales por las que no se usa Internet")
plt.xlabel("Porcentaje de personas (%)")
plt.ylabel("Razón")

plt.tight_layout()
plt.show()
```
