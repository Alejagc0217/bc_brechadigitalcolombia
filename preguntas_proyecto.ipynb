```python
import pandas as pd

# Definir ruta base de tu carpeta
ruta_base = r"C:\Users\Asus\Documents\bc_brechadigitalcolombia\datos"

# 1. Cargar df_v1 con solo las columnas requeridas
df_v1 = pd.read_csv(
    ruta_base + r"\V1_datos_vivienda.CSV",
    encoding="utf-8",
    sep=";",
    low_memory=False
)

# Dejar solo las columnas 'DIRECTORIO' y 'P1_DEPARTAMENTO'
df_v1 = df_v1[['DIRECTORIO', 'P1_DEPARTAMENTO']]

# 2. Cargar el catálogo de códigos de departamento
df_codigos = pd.read_csv(
    ruta_base + r"\codigos_departamentos.csv",
    encoding="utf-8",
    sep=","
)

# 3. Hacer merge con la equivalencia de códigos de departamento
df_v1 = df_v1.merge(
    df_codigos[['Código Departamento', 'Nombre Departamento']].drop_duplicates(),
    how="left",
    left_on="P1_DEPARTAMENTO",
    right_on="Código Departamento"
)

# 4. Renombrar la columna resultante
df_v1 = df_v1.rename(columns={"Nombre Departamento": "departamento"})

# 5. Mostrar primeras filas para verificar
display(df_v1.head())

# 6. Opcional: Verificar si quedaron sin asignar departamento
print("Cantidad de filas sin departamento asignado:")
print(df_v1['departamento'].isna().sum())
```

```python
# Contar hogares por departamento y estado de internet
conteo = df_v1.groupby(['departamento', 'internet']).size().reset_index(name='Cantidad')

# Pivotear la tabla para tener columnas separadas
pivot = conteo.pivot(index='departamento', columns='internet', values='Cantidad').fillna(0)

# Renombrar columnas si es necesario
pivot.columns.name = None
pivot = pivot.rename(columns={"Sí": "Con_Internet", "No": "Sin_Internet"})

# Calcular total y porcentaje
pivot["Total_Hogares"] = pivot["Con_Internet"] + pivot["Sin_Internet"]
pivot["Porcentaje_Con_Internet"] = (pivot["Con_Internet"] / pivot["Total_Hogares"]) * 100

# Ordenar por porcentaje de acceso
pivot_sorted = pivot.sort_values("Porcentaje_Con_Internet", ascending=False)

# Top 5 con mayor acceso
top_5_mayor = pivot_sorted.head(5)

# Top 5 con menor acceso
top_5_menor = pivot_sorted.tail(5)

# Mostrar resultados
print("TOP 5 DEPARTAMENTOS CON MAYOR ACCESO A INTERNET:")
print(top_5_mayor[["Porcentaje_Con_Internet"]])

print("\nTOP 5 DEPARTAMENTOS CON MENOR ACCESO A INTERNET:")
print(top_5_menor[["Porcentaje_Con_Internet"]])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Concatenar top 5 mayor y menor acceso, etiquetando el grupo
top_5_mayor["Grupo"] = "Mayor acceso"
top_5_menor["Grupo"] = "Menor acceso"
df_ranking = pd.concat([top_5_mayor, top_5_menor])

# Ordenar por porcentaje
df_ranking = df_ranking.sort_values("Porcentaje_Con_Internet", ascending=True)

# Configurar estilo
sns.set(style="whitegrid")

# Crear figura
plt.figure(figsize=(10,6))
ax = sns.barplot(
    data=df_ranking,
    y=df_ranking.index,
    x="Porcentaje_Con_Internet",
    hue="Grupo",
    palette="pastel"
)

# Añadir etiquetas de porcentaje
for p in ax.patches:
    width = p.get_width()
    ax.text(
        width + 1,
        p.get_y() + p.get_height() / 2,
        f"{width:.1f}%",
        va="center",
        fontsize=10
    )

# Títulos y etiquetas
plt.title("Departamentos con mayor y menor acceso a Internet", fontsize=16, weight="bold")
plt.xlabel("Porcentaje de hogares con Internet")
plt.ylabel("Departamento")
plt.legend(title="Grupo")

# Ajustar diseño
plt.tight_layout()
plt.show()
```

# Usaremos las columnas P1075S1 QUE HACE REFERENCIA A " Internet FIJO"  1 PARA SI Y 2 PARA NO   y la P1075S2 INTERNET MOVIL  1 PARA SI Y 2 PARA NO 

```python
# Seleccionar las columnas necesarias de df_h3
df_internet_detalle = df_h3[['DIRECTORIO', 'P1075S1', 'P1075S2']].copy()

# Eliminar duplicados por si hay varios registros por DIRECTORIO
df_internet_detalle = df_internet_detalle.drop_duplicates(subset=['DIRECTORIO'])

# Unir al df maestro
df_v1 = df_v1.merge(
    df_internet_detalle,
    how='left',
    on='DIRECTORIO'
)

# Mapear los valores 1/2 a texto para mayor claridad
df_v1['internet_fijo'] = df_v1['P1075S1'].map({1: 'Sí', 2: 'No'})
df_v1['internet_movil'] = df_v1['P1075S2'].map({1: 'Sí', 2: 'No'})

# Eliminar las columnas originales si ya no las quieres
df_v1 = df_v1.drop(columns=['P1075S1', 'P1075S2'])

# Mostrar las primeras filas para verificar
display(df_v1.head())
```

```python
# Crear variables booleanas
df_v1['Tiene_Fijo'] = df_v1['internet_fijo'] == 'Sí'
df_v1['Tiene_Movil'] = df_v1['internet_movil'] == 'Sí'

# Cantidad de viviendas con internet fijo
cantidad_fijo = df_v1['Tiene_Fijo'].sum()

# Cantidad de viviendas con internet móvil
cantidad_movil = df_v1['Tiene_Movil'].sum()

# Cantidad de viviendas que tienen ambos
cantidad_ambos = df_v1[(df_v1['Tiene_Fijo']) & (df_v1['Tiene_Movil'])].shape[0]

# Mostrar resultados
print(f"Cantidad de viviendas con Internet Fijo: {cantidad_fijo:,}")
print(f"Cantidad de viviendas con Internet Móvil: {cantidad_movil:,}")
print(f"Cantidad de viviendas con ambos: {cantidad_ambos:,}")
```

```python
import pandas as pd

# Crear lista para almacenar resultados
resumen = []

# Recorrer cada actividad
for actividad in actividades:
    # Tabla de frecuencias absolutas
    tabla = pd.crosstab(df_p3["Genero"], df_p3[actividad])
    # Tabla de proporciones (normalizado por fila)
    tabla_prop = tabla.div(tabla.sum(axis=1), axis=0) * 100
    # Convertir a DataFrame largo
    tabla_prop = tabla_prop.reset_index().melt(id_vars="Genero", value_name="Porcentaje", var_name="Uso")
    tabla_prop["Actividad"] = actividad
    resumen.append(tabla_prop)

# Concatenar todo en un solo DataFrame
df_resumen = pd.concat(resumen, ignore_index=True)

# Verificar
display(df_resumen.head())
```

```python
# Crear rangos de edad de 0-9, 10-19, ..., 100-109
bins = list(range(0, 120, 10))  # hasta 110, se extiende a 120 por seguridad
labels = [f"{i}-{i+9}" for i in bins[:-1]]

df_p3["Rango_Edad"] = pd.cut(
    df_p3["EDAD"],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True
)

# Verificar
display(df_p3[["EDAD", "Rango_Edad"]].head(10))
```

```python
# Lista de actividades
actividades = [
    'Enviar_correos', 'Redes_sociales', 'Comprar_productos',
    'Banca_electronica', 'Educacion_aprendizaje', 'Tramites_gobierno',
    'Descargar_contenidos', 'Consulta_medios', 'Ver_contenido_audiovisual',
    'Buscar_trabajo', 'Servicios_nube', 'Vender_productos',
    'Trabajo_empleo_independiente', 'Llamadas_videollamadas',
    'Publicar_opiniones', 'Subir_contenido_propio'
]

# Crear lista para almacenar resultados
resumen_edad = []

for actividad in actividades:
    # Contar proporción de uso por rango
    tabla = pd.crosstab(df_p3["Rango_Edad"], df_p3[actividad], normalize="index") * 100
    tabla = tabla.reset_index().melt(id_vars="Rango_Edad", var_name="Uso", value_name="Porcentaje")
    tabla["Actividad"] = actividad
    resumen_edad.append(tabla)

# Concatenar
df_resumen_edad = pd.concat(resumen_edad, ignore_index=True)

# Ver ejemplo
display(df_resumen_edad.head())
```

```python
# Función que asigna grupo cualitativo
def clasificar_edad(edad):
    if edad <= 11:
        return "Niño"
    elif edad <= 17:
        return "Adolescente"
    elif edad <= 29:
        return "Joven adulto"
    elif edad <= 59:
        return "Adulto"
    else:
        return "Adulto mayor"

# Crear nueva columna en df_p3
df_p3["Grupo_Edad"] = df_p3["EDAD"].apply(clasificar_edad)

# Verificar
display(df_p3[["EDAD", "Grupo_Edad"]].head(10))
```

8. ¿Cuál es la razón principal por la que las personas no usan Internet?

```python
# Renombrar columna P1929 por "Razon_no_usa_internet"
df_p3 = df_p3.rename(columns={"P1929": "Razon_no_usa_internet"})
df_p3.columns

```

```python
# Filtrar solo personas que respondieron la razón por la que no usan Internet
df_no_usa_internet = df_p3[df_p3["Razon_no_usa_internet"].notna()].copy()

# Diccionario de códigos de respuesta
razones_dict = {
    1: "Muy costoso",
    2: "No lo necesita",
    3: "No sabe usarlo",
    4: "No hay cobertura en la zona",
    5: "No le permiten usarlo",
    6: "Seguridad o privacidad",
    7: "Otra razón"
}

# Convertir a entero y mapear
df_no_usa_internet["Razon_no_usa_internet"] = df_no_usa_internet["Razon_no_usa_internet"].astype(int).map(razones_dict)

# Calcular frecuencias y porcentajes
conteo = df_no_usa_internet["Razon_no_usa_internet"].value_counts()
porcentaje = (conteo / conteo.sum() * 100).round(2)

# Crear tabla
resumen_razones_final = pd.DataFrame({
    "Razón": conteo.index,
    "Cantidad de personas": conteo.values,
    "Porcentaje (%)": porcentaje.values
})

# Mostrar tabla
print(resumen_razones_final)
```

```python
#Grafico ¿Cuál es la razón principal por la que las personas no usan Internet?
resumen_razones_final = resumen_razones_final.sort_values("Porcentaje (%)", ascending=False)

# Estilo con fondo blanco y colores variados
sns.set(style="whitegrid")

# Crear gráfico con diferentes colores (por ejemplo Set2)
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x="Porcentaje (%)",
    y="Razón",
    data=resumen_razones_final,
    palette="Paired",
    orient="h"
)

# Etiquetas al final de cada barra
for index, value in enumerate(resumen_razones_final["Porcentaje (%)"]):
    plt.text(value + 0.5, index, f"{value}%", va='center')

# Títulos
plt.title("Razones principales por las que no se usa Internet")
plt.xlabel("Porcentaje de personas (%)")
plt.ylabel("Razón")

plt.tight_layout()
plt.show()
```
