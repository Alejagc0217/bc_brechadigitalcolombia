# Quienes usan interneto y computador? 

# PAra esto usamos coilumnas "P1084" "P1910"

# Renombramos "P1084" como "usa_internet"

# P1910" Usa computador" 

```python
# Crear variable de acceso a Internet
# 1 si usa Internet en cualquier frecuencia (1 a 4), 0 si no usa (5) o NaN si falta dato
df_2024["usa_internet"] = df_2024["P1084"].apply(lambda x: 1 if x in [1,2,3,4] else (0 if x==5 else pd.NA))

# Crear variable de acceso a computador (escritorio o portátil)
df_2024["usa_computador"] = df_2024.apply(
    lambda row: 1 if (row["P1910"] in [1,2,3,4] or row["P1911"] in [1,2,3,4]) else (
        0 if (row["P1910"]==5 and row["P1911"]==5) else pd.NA
    ),
    axis=1
)

# Verificar primeras filas
df_2024[["P1084", "P1910", "P1911", "usa_internet", "usa_computador"]].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set(style="whitegrid", palette="muted")

# Preparar datos agregados ponderados
# Peso muestral
peso = df_2024["FEX_C"]

# Calcular porcentajes ponderados
uso_internet = df_2024["usa_internet"].value_counts(normalize=False)
uso_computador = df_2024["usa_computador"].value_counts(normalize=False)

# Porcentajes ponderados
uso_internet_pond = df_2024.groupby("usa_internet")["FEX_C"].sum() / peso.sum() * 100
uso_computador_pond = df_2024.groupby("usa_computador")["FEX_C"].sum() / peso.sum() * 100

# Crear DataFrames para graficar
df_internet_plot = uso_internet_pond.reset_index()
df_internet_plot.columns = ["Acceso Internet", "Porcentaje"]

df_computador_plot = uso_computador_pond.reset_index()
df_computador_plot.columns = ["Acceso Computador", "Porcentaje"]

# Mapeo de etiquetas
etiquetas = {1: "Sí", 0: "No"}

df_internet_plot["Acceso Internet"] = df_internet_plot["Acceso Internet"].map(etiquetas)
df_computador_plot["Acceso Computador"] = df_computador_plot["Acceso Computador"].map(etiquetas)

# Crear figura
fig, axes = plt.subplots(1, 2, figsize=(14,6))

# Gráfico de acceso a Internet
sns.barplot(
    data=df_internet_plot,
    x="Acceso Internet",
    y="Porcentaje",
    ax=axes[0]
)
axes[0].set_title("Porcentaje de personas con acceso a Internet")
axes[0].set_ylabel("Porcentaje")
axes[0].set_ylim(0, 100)
for i, v in enumerate(df_internet_plot["Porcentaje"]):
    axes[0].text(i, v + 1, f"{v:.1f}%", ha="center")

# Gráfico de acceso a Computador
sns.barplot(
    data=df_computador_plot,
    x="Acceso Computador",
    y="Porcentaje",
    ax=axes[1]
)
axes[1].set_title("Porcentaje de personas con acceso a Computador")
axes[1].set_ylabel("Porcentaje")
axes[1].set_ylim(0, 100)
for i, v in enumerate(df_computador_plot["Porcentaje"]):
    axes[1].text(i, v + 1, f"{v:.1f}%", ha="center")

plt.tight_layout()
plt.show()
```
